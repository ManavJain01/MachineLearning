```python
df=pd.DataFrame(df,columns=['review','score'])
df
```

```python
#Text Cleaning

# 1.Sample 10000 rows
# 2.Remove html tags
# 3.Remove special characters
# 4.Converting every thing to lower case
# 5.Removing Stop words
# 6.Stemming
```

```python
# df=df.sample(10000)
df=df.sample(1000)
```

```python
df.info()
```

```python
df
```

```python
import re
clean = re.compile('<.*?>')
re.sub(clean,'',df.iloc[1].review)
```

```python
# converting everything to lower
def convert_lower(text):
    return text.lower()

df['review']=df['review'].apply(convert_lower)
df
```

```python
# function to remove special characters
def remove_special(text):
    x=''
    
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x + ' '
    return x

df['review']=df['review'].apply(remove_special)
df
```

```python
#Remove the stop words
import nltk
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear()
    return y

df['review']=df['review'].apply(remove_stopwords)
df
```

```python
vg_stopwords=['videogame','game','play','player','character','tlou','tlou2', 'neil','druckmann','naughty','dog','nc',
 'aaa', 'na','youtube', 'youtuber','f','h','ng']

def remove_VGstopwords(text):
    x=[]
    for i in text:
        if i not in vg_stopwords:
            x.append(i)
    y=x[:]
    x.clear()
    return y

df['review']=df['review'].apply(remove_VGstopwords)
df
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z

df['review']=df['review'].apply(stem_words)
df
```

```python
# Join back
def join_back(list_input):
    return " ".join(list_input)

df['review'] = df['review'].apply(join_back)
df['review']
```
