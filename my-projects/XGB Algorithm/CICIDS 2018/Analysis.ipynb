```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('./Dataset/02-14-2018.csv')
df2 = pd.read_csv('./Dataset/03-02-2018.csv')
df3 = pd.read_csv('./Dataset/02-15-2018.csv')
# df1 = (1048575, 80)
# df2 = (1048575, 80)
# df3 = (1048575, 80)

df1 = df1.sample(n=500000)
df2 = df2.sample(n=500000)
df3 = df3.sample(n=500000)
```

```python
display(df1.shape, df2.shape, df3.shape)
```

```python
df = pd.concat([df1,df2,df3])
df
```

```python
del df1
del df2
del df3
```

```python
total = df.shape[0]
missing_columns = [col for col in df.columns if df[col].isnull().sum()>0]
for col in missing_columns:
    null_count = df[col].isnull().sum()
    per = (null_count/total) * 100
    print(f"{col}: {null_count} ({round(per, 3)}%)")
```
