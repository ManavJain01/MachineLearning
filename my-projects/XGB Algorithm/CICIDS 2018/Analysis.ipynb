```python
df = pd.concat([df1,df2,df10])

del df1
del df2
del df10

df
```

# Data Visualization

```python
# the column 'Flow Byts/s' has 4707 null values and 1495296 not-null values
df['Flow Byts/s'].isnull().sum()
```

```python
total = df.shape[0]
missing_columns = [col for col in df.columns if df[col].isnull().sum()>0]
for col in missing_columns:
    null_count = df[col].isnull().sum()
    per = (null_count/total) * 100
    print(f"{col}: {null_count} ({round(per, 3)}%)")
```
